```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
df = sns.load_dataset('flights')
df.head(20)
```

# ***RELATIONAL PLOTS***

```python
sns.relplot(x = 'passengers',y = 'month',hue = 'year',data = df)
```

```python
sns.relplot(x = 'passengers',y = 'month',hue = 'year',data = df,kind = 'line')
```

```python
df = sns.load_dataset('tips')
df.head()
```

# ***CATEGORICAL PLOTS***

```python
sns.catplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'violin')
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'swarm')
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'box')
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'boxen')
```

# ***DISTRIBUTION PLOT***

```python
sns.distplot(df['total_bill'],bins = 30)
```

```python

sns.jointplot(x ='total_bill', y ='tip', data = df)
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df, kind ='reg')
```

```python
sns.jointplot(x ='total_bill', y ='tip', data = df, kind ='kde')
```
