```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
df.head(20)
```

# ***RELATIONAL PLOTS***

```python
sns.relplot(x = 'passengers',y = 'month',hue = 'year',data = df)
```

```python
sns.relplot(x = 'passengers',y = 'month',hue = 'year',data = df,kind = 'line')
```

```python
df = sns.load_dataset('tips')
df.head()
```

# ***CATEGORICAL PLOTS***

```python
sns.catplot(x = 'day',y = 'total_bill',data = df)
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'violin')
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'swarm')
```

```python
sns.catplot(x = 'day',y = 'total_bill',data = df,kind = 'box')
```
